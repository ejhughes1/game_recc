```python
col_names = ['steamid','appid','playtime_2weeks','playtime_forever','dateretrieved']
gamers_df = pd.read_csv('../data/Games_1_subset.csv',quotechar="'",names=col_names)
```

```python
gamers_df
```

```python
top10_df = gamers_df.groupby('appid').agg({'playtime_forever':'mean', 'dateretrieved':'count'}).sort_values('playtime_forever')
```

```python
top10_df
```

```python
joined_df = top10_df.merge(games_df, left_index = True, right_on = 'appid')
```

```python
joined_df.sort_values('playtime_forever', ascending = False).head(50)
```

```python
joined_df['playtime_hours'] = joined_df['playtime_forever']/60
```

```python
joined_df.sort_values('playtime_forever', ascending = False)[['playtime_forever', 'dateretrieved', 'title', 'playtime_hours']].head(50)
```
